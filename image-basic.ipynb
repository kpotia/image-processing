# Loading images with imageio

we use imageio to have more flexibility.

```python
import imageio.v3 as iio

eight = iio.imread(uri="data/eight.tif")
print(type(eight))
```

# Working with pixels

```python
import ipympl
import matplotlib.pyplot as plt # load pyplot as plt 
import numpy as np
import skimage as ski

%matplotlib widget

fig, ax = plt.subplots()
ax.imshow(eight)
```

```python
print(eight.shape)
print(eight)
```

```python
zero = iio.imread(uri='data/eight.tif')
zero[2,1] = 1.0

# The following line of code creates a new figure for imshow to use in displaying our output.
fig, ax = plt.subplots()
ax.imshow(zero)
print(zero)
```

```python
five = iio.imread(uri='data/eight.tif')
five[1,2] = 1.0
five[3,0] = 1.0
# five[,2] = 1.0

# The following line of code creates a new figure for imshow to use in displaying our output.
fig, ax = plt.subplots()
ax.imshow(five)
print(five)
```

```python
# make a copy of eight 
three_colours = iio.imread("data/eight.tif")

# multiply the whole matrix by 128
three_colours = three_colours * 128

# set the middle row(index 2) the value of 255.,
# so you end up with the values 0., 128. and 255. 

three_colours[2,:] = 255.
fig, ax = plt.subplots()
ax.imshow(three_colours)
print(three_colours)

```

```python
fig, ax = plt.subplots()
ax.imshow(three_colours, cmap=plt.cm.gray)
```

```python
# set the random seed so we all get the same matrix
pseudorandomizer = np.random.RandomState(2021)

#create a 4x4 checkboard of random colours 
checkboard = pseudorandomizer.randint(0,255, size=(4,4,3))

#restore the default map as you show the image 
fig, ax = plt.subplots()

ax.imshow(checkboard)
print(checkboard)
```
